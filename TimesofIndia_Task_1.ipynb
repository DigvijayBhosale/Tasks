# Code to navigate 2018's archive news Month wise

```python

news = driver.find_elements_by_xpath("//a[@class='normtxt']")
links=[]
for i in news:
    url=i.get_attribute('href')
    match=re.search('year-2018',url)
    if match:
        links.append(url)
for i in links:
    mon=re.search('year-2018,month-13',i)
    if mon:
        links.remove(i)
links
```

# Code to consider only Aug Month's data from 2018

```python
for i in links:
    aug=re.search('year-2018,month-8',i)
    if aug:
        driver.get(i)
```

# Code will append links of Aug's every date to calnde list

```python
calndr = []
for i in range(6):
    i = i+1
    path = ("//*[@id='calender']/tbody/tr["+str(i)+"]")
    for i in range(7):
        i += 1
        calnd = (path+"/td["+str(i)+"]/a")
        a = driver.find_elements_by_xpath(calnd)
        for i in a:
            if i.text:                
                url=i.get_attribute('href')
                print(url)
                calndr.append(url)
        
        
```

# This code will navigate to each date of Aug month of 2018 and will scrap the links of all the news repoarted on that day

```python
news = []
for i in calndr:
    driver.get(i)
    time.sleep(3)
    nws = driver.find_elements_by_xpath("/html/body/div[1]/table[2]/tbody/tr[2]/td[1]/div[3]/table/tbody/tr[2]/td[1]/span/a")
    for i in nws:
        daynews = i.get_attribute('href')
        news.append(daynews)
```
